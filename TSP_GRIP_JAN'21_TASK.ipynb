```
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
%matplotlib inline
```

```
url= 'http://bit.ly/w-data'
data= pd.read_csv(url)
print("The data is imported sucessfully")
data.head(25)

```

```
data.describe()
```

Now we have to visualize the relation between scores and study hour: 

*   independent variable: study hour
*   dependent hour: score




```
data.plot(x='Hours', y='Scores', style='o')
plt.title('score and hour studied')
plt.xlabel('Hours studied')
plt.ylabel('score')
plt.show()
```

```
x=data.iloc[:, :-1].values
y=data.iloc[:, 1:].values
```

 Now we have the values (x,y) so we have to split these into training and testing dataset, to do so we are using train_test_split function from sklearn.model_selection

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=0)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
regr = LinearRegression()
regr.fit(x_train, y_train)
print("Training dataset")
print(regr.coef_)
print(regr.intercept_)
```

Now we have to make the regression line and plot it on the graph with the dataset



```
#plotting the line
line= regr.coef_*x + regr.intercept_
plt.scatter(x,y)
plt.plot(x, line)
plt.show()



```
